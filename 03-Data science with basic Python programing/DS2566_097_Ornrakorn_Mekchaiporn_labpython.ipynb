### **Exercise 3: Function**

จงสร้างฟังก์ชัน convert_month ในการแปลงเดือนจากตัวเลข เป็นชื่อเดือน โดยกำหนดให้ input อยู่ในรูปแบบของ dd-mm-yyyy

Screenshot 2023-07-09 200733.png

```python
def convert_month(date_input):
  date = date_input.split('-')

  day = date[0].lstrip('0')
  #print(day)

  month = date[1].lstrip('0')
  #print(month)

  year = date[2].lstrip('0')
  #print(year)

  if month == '1':
    print(day,'มกราคม',year)

  elif month == '2':
      print(day,'กุมภาพันธ์',year)

  elif month == '3':
      print(day,'มีนาคม',year)

  elif month == '4':
      print(day,'เมษายน',year)

  elif month == '5':
      print(day,'พฤษภาคม',year)

  elif month == '6':
      print(day,'มิถุนายน',year)

  elif month == '7':
      print(day,'กรกฎาคม',year)

  elif month == '8':
      print(day,'สิงหาคม',year)

  elif month == '9':
      print(day,'กันยายน',year)

  elif month == '10':
      print(day,'ตุลาคม',year)

  elif month == '11':
      print(day,'พฤศจิกายน',year)

  elif month == '12':
      print(day,'ธันวาคม',year)

  else:
    print('รูปแบบของวันที่ใส่มาไม่ถูกต้อง')


convert_month('0031-01-002500')
convert_month('19-05-2564')
convert_month('09-09-2549')
convert_month('31-012-2500')
```

```python
# 2.Creating DataFrame from a Dictionary
d1 = {'Name': ['Annie', 'Tom','Eddy'], 'Gender': ['Female', 'Male', 'Male'], 'Age':[23, 57, 35]}
df = pd.DataFrame(d1)
df
```

```python
# 3.Creating DataFrame from a List
a_list = [['Annie', 'Female', 23], ['Tom', 'Male', 57], ['Eddy', 'Male', 35]]
df = pd.DataFrame(a_list,columns=['Name','Gender','Age'])
df
```
